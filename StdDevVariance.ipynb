# Standard Deviation and Variance

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def plot_normal(mean, std, number_of_points):
    array = np.random.normal(mean, std, number_of_points)

    computed_mean = array.mean()
    computed_std = array.std()
    computed_variance = array.var()
    plt.hist(array, 50)
    plt.axvline(x=computed_mean, color='green')
    plt.axvline(x=computed_mean+computed_std, color='red')
    plt.axvline(x=computed_mean-computed_std, color='red')
    plt.show()
    print(computed_mean, computed_std, computed_variance)
    
plot_normal(100, 50, 10000)
```
